```python
from ipywidgets import IntSlider
```

```python
slider = IntSlider(value=50)

slider
```

```python
slider.value = 30
```

```python
from ipywidgets import HTML
```

```python
text = HTML("The silder value is <b>{}</b>".format(slider.value))

text
```

```python
def update_html(change):
  text.value = "The silder value is <b>{}</b>".format(slider.value)

# Execute callback whenever the slider value is changed
slider.observe(update_html, 'value')
```

```python
import cv2
import matplotlib

from skimage import io, img_as_float
from skimage.filters import gaussian
```

```python
pixels = cv2.imread('woman2.jpg')
```

```python
matplotlib.pyplot.imshow(pixels)
```

```python
filter_cv2 = cv2.GaussianBlur(pixels, (5,5), 0, borderType=cv2.BORDER_CONSTANT)
```

```python
filter_skimage = gaussian(pixels, sigma = 5, mode = 'constant', cval = 0.0)
```

```python
matplotlib.pyplot.imshow(filter_cv2)
```

```python
matplotlib.pyplot.imshow(filter_skimage)
```
